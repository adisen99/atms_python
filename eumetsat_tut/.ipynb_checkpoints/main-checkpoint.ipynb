```python
no2 = xr.open_dataarray("./data/download.nc")
no2
```

### Extracting and Plotting the point data

```python
first_point = no2.isel(latitude=0,longitude=0)
area = no2.sel(latitude=50.05,longitude=8.05)
area.plot()
```

```python
no2.sel(latitude=50.05,longitude=[1.05, 8.05],level=0.0).plot.line(x = "time")
```

### Histograms

We can also plot histograms of the entire data set and in order to have more sensible looking bins we plot the logarithm of the no2 data

```python
np.log(no2).plot()
```

## Maps

Visualizing the data on a map

```python
# Using the geophysical units and 'robust' disregards outliers for the colormap creation

no2.sortby(['longitude','latitude'],ascending=True).isel(time=0).plot(robust=True)
```

```python
# Centered over Darmstadt

ax = plt.axes(projection=ccrs.EuroPP())
ax.coastlines()

# Use the last time step in array
p = no2.sortby(['longitude','latitude'],ascending=True).isel(time=-4).plot(transform=ccrs.PlateCarree(),robust=True)
```
